```python
#print(res)
#print(len(res.cumcount))
left=0
#print(res)
num_cluster=1
df_results_clean.loc[:,'cluster']=0
feature_average=[]
for i in range(0, len(res.cumcount)):
    right=int(res.cumcount[i])
    cluster=df_results_clean[left:right].copy()
    df_results_clean.loc[left:right, 'cluster']=num_cluster
    num_cluster=num_cluster+1
    lstc=list(cluster['avg'])
    #print('# cluster', i, 'from', left, 'to', right, 'with', len(lstc), 'elements')
    #print(len(lstc))
    if(len(lstc) > 0):
        #print('## range from', lstc[0], 'to', lstc[-1])
        columns=list(cluster.filter(regex='.+\..+', axis=1))
        local_res={}
        for c in columns:
            avgr=sum(map(lambda x: int(x), list(cluster[c])))/len(lstc)
            local_res[c]= avgr
            #print('-', c, avgr)
        feature_average.append(local_res)
    left=right
#print(df_results_clean['cluster'])
favg = pd.DataFrame.from_dict(feature_average)
print(favg.describe(include='all'))
print(tabulate(favg, tablefmt="pipe", headers="keys"))
```
